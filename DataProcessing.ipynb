```python
import pandas as pd
import datetime
import sys
```

```python
df1=pd.read_csv("Jan2016-2018_1000.csv",header=0)
```

```python
df1
df1['Total'] = df1.groupby(['localminute'])['dataid'].transform('sum')
df1['DateTime'] = [datetime.datetime.strptime(d, "%Y-%m-%d %H:%M:%S-%f") for d in df1["localminute"]]
df1['DateTime'] = df1['DateTime'].dt.floor('h')
```

```python
tmp = df1.groupby('localminute')['dataid'].apply(set)
a0 = tmp[0] & tmp[0]
for i in tmp:
    a0 = a0 & i
print (a0)
df = df1[df1['dataid'].isin(a0)]
df['Total1'] = df.groupby(['localminute'])['dataid'].transform('sum')
```

```python
df['Total1'].unique() #Should return exactly value. Total1 <= min(df1.Total)
```

```python
del df['dataid']
df['TotalUsage'] = df.groupby(['DateTime'])['use'].transform('sum')
df = df.drop_duplicates(subset='DateTime')
df = df.reset_index(drop=True)
```

```python
df['Date']= df['DateTime'].dt.date
df['Time']= df['DateTime'].dt.time
df['Hour'] = df['DateTime'].dt.hour
df['Minute'] = df['DateTime'].dt.minute
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Year'] = df['DateTime'].dt.year
```
